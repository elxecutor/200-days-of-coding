```python
import math
import matplotlib.pyplot as plt

# Generate values from 1 to 100
n = list(range(1, 101))

# Complexity functions
o1 = [1 for _ in n]
ologn = [math.log(x) for x in n]
on = [x for x in n]
onlogn = [x * math.log(x) for x in n]
on2 = [x**2 for x in n]
on_k = [x**3 for x in n]

# Plotting each complexity function
plt.figure(figsize=(10, 6))
plt.plot(n, o1, label="O(1)")
plt.plot(n, ologn, label="O(log n)")
plt.plot(n, on, label="O(n)")
plt.plot(n, onlogn, label="O(n log n)")
plt.plot(n, on2, label="O(n²)")
plt.plot(n, on_k, label=f"O(n^{k_val})")

plt.xlabel("Input Size (n)")
plt.ylabel("Time Complexity (arbitrary units)")
plt.title("Comparison of Time Complexity Classes")
plt.legend()
plt.grid(True)
plt.ylim(0, 100)
plt.xlim(0, 100)
plt.show()
```
